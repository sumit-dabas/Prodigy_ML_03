Implement a support vector machine (SVM) to classify images of cats and dogs from the Kaggle dataset.

Dataset :- https://www.kaggle.com/c/dogs-vs-cats/data

```python
import numpy as np
import os
import cv2
import joblib
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score
```

```python
# Define paths based on your folder structure
base_path = "C:/Users/Sumit Dabas/My Projects/Prodigy ML Internship/Prodigy_ML_03/PetImages"
cat_folder = os.path.join(base_path, 'Cat')
dog_folder = os.path.join(base_path, 'Dog')
```

```python
print(cat_folder)
print(dog_folder)
```

```python
# Image preprocessing settings
image_size = (64, 64)  # Resize images to 64x64
data = []
labels = []
```

```python
def preprocess_images(folder, label):
    for img_name in os.listdir(folder):
        img_path = os.path.join(folder, img_name)
        
        # Only process images with valid extensions
        if img_name.endswith('.jpg') or img_name.endswith('.jpeg') or img_name.endswith('.png'):
            try:
                # Read the image and convert to grayscale
                img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
                img = cv2.resize(img, image_size)  # Resize image
                data.append(img.flatten())  # Flatten the image into a 1D array
                labels.append(label)  # Append corresponding label
            except Exception as e:
                print(f"Error loading image {img_path}: {e}")
```

```python
def preprocess_images(folder, label):
    for img_name in os.listdir(folder):
        img_path = os.path.join(folder, img_name)
        
        # Only process images with valid extensions
        if img_name.endswith('.jpg') or img_name.endswith('.jpeg') or img_name.endswith('.png'):
            try:
                # Read the image and check if it was successfully loaded
                img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
                
                if img is None:
                    print(f"Failed to load image: {img_path}")
                    continue
                
                img = cv2.resize(img, image_size)  # Resize image
                data.append(img.flatten())  # Flatten the image into a 1D array
                labels.append(label)  # Append corresponding label
            except Exception as e:
                print(f"Error processing image {img_path}: {e}")
```

```python
# Preprocess cat and dog images
preprocess_images(cat_folder, 0)  # Label 0 for cats
preprocess_images(dog_folder, 1)  # Label 1 for dogs
```

```python
# Convert data and labels to numpy arrays
data = np.array(data)
labels = np.array(labels)
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, random_state=42)
```

```python
# Standardize the data (normalize features)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Create and train the SVM model
svm_model = SVC(kernel='linear', random_state=42)
svm_model.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = svm_model.predict(X_test)
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
# Print a detailed classification report
print("Classification Report:\n", classification_report(y_test, y_pred))
```
